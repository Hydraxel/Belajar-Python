```python
for i in range(1, 14, 2):
    if i == 7:
        print("wow")
    else:
        print(i)
```


Analisis :
pada code ini melakukan perulangan untuk mencetak angka ganjil dari 1 hingga 13, dengan menggunakan range(1, 14, 2), yang menghasilkan angka 1, 3, 5, 7, 9, 11, dan 13. Jika angka yang sedang diproses adalah 7, maka akan mencetak kata "wow", sedangkan untuk angka lainnya, program akan mencetak angka tersebut.

```python
for i in range(1, 14, 2):
    if i == 7:
        print("wow")
    else:
        print(i)
```


Analisis : 
pada code ini melakukan perulangan untuk mencetak angka ganjil dari 1 hingga 13, dimulai dari 1 dengan langkah 2 (jadi angka yang dicetak adalah 1, 3, 5, 7, 9, 11, 13). Namun, jika nilai i adalah 7, maka yang dicetak adalah kata "wow" daripada angka 7. Untuk nilai i yang lainnya, program akan mencetak angka tersebut.

```python
i = 1
while i < 14:
    if i == 7:
        print("wow")
    else:
        print(i)
    i += 2
```


Analisis : 
 Pada code ini menggunakan perulangan while untuk mencetak angka ganjil mulai dari 1 hingga kurang dari 14. Pada setiap iterasi, jika nilai i sama dengan 7, maka akan mencetak "wow", jika tidak, mencetak nilai i. Setelah setiap iterasi, nilai i akan bertambah 2, sehingga hanya angka ganjil yang akan dicetak (1, 3, 5, 7, 9, 11, 13).

```python
i = 1
while i < 11:
    if i == 3:
        print("ha")
    else:
        print(i)
    
    if i == 5:
        print("ha")
    
    if i == 8:
        print("ha")
    
    i += 1
```
