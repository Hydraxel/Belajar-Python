Varibel global
Variabel yang dibuat di luar fungsi (seperti dalam semua contoh di halaman sebelumnya) dikenal sebagai variabel global.

Variabel global dapat digunakan oleh siapa saja, baik di dalam fungsi maupun di luar.

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

Jika kita membuat variabel dengan nama yang sama di dalam suatu fungsi, variabel ini akan bersifat lokal dan hanya dapat digunakan di dalam fungsi tersebut. Variabel global dengan nama yang sama akan tetap sama, global, dan dengan nilai aslinya.

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

 Pada Kata Kunci Global
 Biasanya, ketika kita membuat variabel di dalam suatu fungsi, variabel tersebut bersifat lokal, dan hanya dapat digunakan di dalam fungsi tersebut.
Untuk membuat variabel global di dalam suatu fungsi, kita dapat menggunakan kata global sebagai kata kunci.

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

Gunakan juga kata global sebagai kata kunci jika anda ingin mengubah varibael global di dalam suatu funsi. Untuk mengubah nilai varibel di dalam suatu fungsi, rujuk varibel tersebut dengan menggunakan global sebagai kata kunci.

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

Python Numbers

ada tiga tipe data numerik dalam python:
int,
float dan 
complex.
varibel bertipe numerik dibuat saat anda menetapkan nilai padanya.


```python
x = 1    # int
y = 2.8  # float
z = 1j   # complex
```

Untuk memverifikasi jenis objek apa pun dalam Python, gunakan _type()_

1. int (Integer): Tipe data ini digunakan untuk menyimpan bilangan bulat (tanpa desimal). Contoh: x = 1 adalah sebuah bilangan bulat.

2. float (Floating Point): Tipe data ini digunakan untuk menyimpan angka dengan desimal atau pecahan. Contoh: y = 2.8 adalah sebuah angka desimal.

3. complex (Kompleks): Tipe data ini digunakan untuk menyimpan bilangan kompleks, yang terdiri dari bagian real dan bagian imajiner. Bilangan kompleks ditulis dengan format a + bj, di mana a adalah bagian real dan b adalah bagian imajiner. Contoh: z = 1j adalah bilangan kompleks dengan bagian real 0 dan bagian imajiner 1.

```python
x = 1
y = 2.8
z = 1j

print(type(x))
print(type(y))
print(type(z))
```

Semua string adalah True, kecuali string kosong.

Operator Walrus
Python 3.8 memperkenalkan operator := , yang dikenal sebagai "operator walrus". Operator ini menetapkan nilai ke variabel sebagai bagian dari ekspresi yang lebih besar:

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

Operator Bitwise
Operator bitwise digunakan untuk membandingkan angka (biner).

Operator & membandingkan setiap bit dan menetapkannya ke 1 jika keduanya bernilai 1, jika tidak maka ditetapkan ke 0:

Representasi biner dari 6 adalah 0110
Representasi biner dari 3 adalah 0011

Kemudian operator & membandingkan bit dan mengembalikan 0010, yang merupakan 2 dalam desimal.

```python
print(6 & 3)
```

Contoh
Operator | membandingkan setiap bit dan menetapkannya ke 1 jika satu atau keduanya bernilai 1, jika tidak maka ditetapkan ke 0:

print(6 | 3)
Representasi biner dari 6 adalah 0110
Representasi biner dari 3 adalah 0011

Kemudian operator | membandingkan bit dan mengembalikan 0111, yang merupakan 7 dalam desimal.

```python
print(6 | 3)
```

Contoh
Operator ^ membandingkan setiap bit dan menetapkannya ke 1 jika hanya satu yang bernilai 1, jika tidak (jika keduanya bernilai 1 atau keduanya bernilai 0) ditetapkan ke 0:

print(6 ^ 3)
Representasi biner dari 6 adalah 0110
Representasi biner dari 3 adalah 0011

Kemudian operator ^ membandingkan bit dan mengembalikan 0101, yang merupakan 5 dalam desimal.

```python
print(6 ^ 3)
```
